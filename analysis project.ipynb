```python
data = pd.read_csv("marketing_campaign.csv", sep="\t")
data
```

```python
#handling null values
data['Income'] = data['Income'].astype(str)
data['Income'] = data['Income'].str.replace('[\$,]', '', regex=True).str.strip()
data['Income'] = pd.to_numeric(data['Income'], errors='coerce')
data['Income'] = data['Income'].fillna(data['Income'].median())
data['Income']
```

```python
data['Income'].isnull().sum()
```

```python
#ensure no any duplicates
data.duplicated().sum()
```

```python
#remove unusful columnns
data.drop(columns=['Z_CostContact','Z_Revenue'],inplace=True)
```

```python
data.head(2)
```

```python
#convert Dt_Customer column to date
data['Dt_Customer']=pd.to_datetime(data['Dt_Customer'],dayfirst=True)
#add new column to now old,new customers
data['Customer_Since_Days'] = (pd.Timestamp('today') - data['Dt_Customer']).dt.days
data.head(5)
```

```python
# تنظيف القيم المفقودة
data.dropna(subset=["Income", "Response"], inplace=True)
```

```python
data['age']=pd.Timestamp('today').year-data['Year_Birth']
```

```python
data.head(3)
```

```python
data['Marital_Status'] = data['Marital_Status'].replace({
    'Together': 'Married',
    'Divorced': 'Single',
    'Widow': 'Single',
    'Alone': 'Single',
    'YOLO': 'Single',
    'Absurd':'Single'
})
```

```python
data['Marital_Status'].unique()
```

```python
outlier_cols = [
    'Year_Birth', 'Income', 'MntWines', 'MntFruits', 'MntMeatProducts',
    'MntFishProducts', 'MntSweetProducts', 'MntGoldProds',
    'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases',
    'NumWebVisitsMonth'
]

#before
plt.figure(figsize=(15, 40))
for i, col in enumerate(outlier_cols, 1):
    plt.subplot(len(outlier_cols), 1, i)
    sns.boxplot(x=data[col], color='skyblue')
    plt.title(f'Before Cleaning - {col}')
    plt.tight_layout()
plt.show()

# remove outliers
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
   
for col in outlier_cols:
    data = remove_outliers_iqr(data, col)

# after
plt.figure(figsize=(15, 40))
for i, col in enumerate(outlier_cols, 1):
    plt.subplot(len(outlier_cols), 1, i)
    sns.boxplot(x=data[col], color='lightgreen')
    plt.title(f'After Cleaning - {col}')
    plt.tight_layout()
plt.show()
```

EDA (Exploratory Data Analysis)

```python
# Most Common Age
most_common_age = data['age'].mode()[0]
print("Most Common Age:", most_common_age)
```

```python
# Total Customers
total_customers = data.shape[0]
print("Total Customers:", total_customers)
```

```python
# Average Income
avg_income = data['Income'].mean().round()
print("Average Income:", avg_income)
```

```python
# Total Spend
spend_cols = ['MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds']
data['Total_Spend'] = data[spend_cols].sum(axis=1)
print("Total Spend:", data['Total_Spend'].sum())
print("Average Spend per Customer:", data['Total_Spend'].mean())
```

Spending

```python
# Top Product Category by Spend
top_product = data[spend_cols].mean().idxmax()
print("Top Product Category:", top_product)

# Spend based on product type
print("\nSpend per Product Type:")
print(data[spend_cols].sum().sort_values(ascending=False))
```

Purchase Channels



```python
purchase_channels = ['NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']

# Top Purchase Channel
top_channel = data[purchase_channels].mean().idxmax()
print("Top Channel:", top_channel)

# Purchase Counts per Channel
print("\nPurchase Counts per Channel:")
print(data[purchase_channels].sum().sort_values(ascending=False))

# Website Visits
print("Average Website Visits per Month:", data['NumWebVisitsMonth'].mean())
```

Marketing campaign

```python
cmp_cols = ['AcceptedCmp1','AcceptedCmp2','AcceptedCmp3','AcceptedCmp4','AcceptedCmp5']
data['Campaign_Accepted_Count'] = data[cmp_cols].sum(axis=1)

# Response Rate
response_rate = data['Response'].mean()
print("Response Rate:", response_rate)

# Count of customers who accepted multiple campaigns
print("Customers who accepted >1 campaign:", (data['Campaign_Accepted_Count'] > 1).sum())

# Number of responses per campaign
print("\nResponses per Campaign:")
print(data[cmp_cols].sum().sort_values(ascending=False))
```

Customer Segmentation

```python
# Old vs New Customers
data['Customer_Since_Days'] = (pd.to_datetime('today') - data['Dt_Customer']).dt.days
old_customers = (data['Customer_Since_Days'] > 1000).sum()
new_customers = (data['Customer_Since_Days'] <= 1000).sum()

print("Old Customers (>1000 days):", old_customers)
print("New Customers (<=1000 days):", new_customers)
```
